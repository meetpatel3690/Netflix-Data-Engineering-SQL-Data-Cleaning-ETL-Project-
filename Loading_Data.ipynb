In this file We are trying to import .csv data of Netflix and will try to import them to Postgres SQL.

```python
import pandas as pd

raw_netflix_df = pd.read_csv('netflix_titles.csv')

```

Now we will import this .csv file into Postgres DB as a table.

```python
pip install psycopg2-binary sqlalchemy
```

```python
from sqlalchemy import create_engine

# Step 1 : Connect to the PostgreSQL server
# Replace the placeholders with your actual database credentials

username = 'postgres' # your username
password = '3690' # your password
host = 'localhost' # your host, usually localhost
port = '5432' # your port, usually 5432
database = 'Meet' # your database name

engine = create_engine(f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}')

# Step 2 : Load DataFrame to PostgreSQL

table_name = "Netflix_raw" # name of the table to be created
raw_netflix_df.to_sql(table_name, engine, if_exists='replace', index=False)

print(f"DataFrame loaded successfully into the '{table_name}' table in the '{database}' database.")
```
